# SETTING UP VALIDATION FRAMEWORK

```python
# Split data into train, validation, test
from sklearn.model_selection import train_test_split

df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df_full_train), len(df_train), len(df_val), len(df_test)
```

```python
df_full_train = df_full_train.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train['Class'].values
y_val = df_val['Class'].values
y_test = df_test['Class'].values


del df_train['Class']
del df_val['Class']
del df_test['Class']

df_full = df_full_train.copy()
y_full = df_full['Class']

del df_full['Class']
```

```python
df_full = df_full.copy
df_train_copy = df_train.copy()
df_val_copy = df_val.copy()
df_test_copy = df_test.copy()

y_full_copy = y_full.copy()
y_train_copy = y_train.copy()
y_val_copy = y_val.copy()
y_test_copy = y_test.copy()
```

## BASE MODELS - WITHOUT FEATURE SELECTION OR OVERSAMPLING

```python
models = [
    LogisticRegression(random_state=42),
    SVC(random_state=42, probability=True),
    GaussianNB(),
    RandomForestClassifier(random_state=42),
    XGBClassifier(random_state=42)
    ]

model_names = [
    'Logistic Regression',
    'SVM',
    'Gaussian Naive Bayes',
    'Random Forest',
    'XGBoost'
    ]
```

## TRAINING PIPELINE WITH FEATURE SCALING

### STANDARDIZATION

```python
pipelines = []

for model in models:
  pipe = Pipeline([
      ('scaler', StandardScaler()),
      ('classifier', model)
      ])
  pipelines.append(pipe)
```

```python
pipelines
```

## TRAINING AND EVALUATION ON VALIDATION SET

```python
# Lists to store evaluation metrics across models
all_accuracy = []
all_recall = []
all_precision = []
all_roc_auc = []
all_fpr = []
all_tpr = []
```

```python
# Train and evaluate each model using the pipelines

for model_name, pipeline in zip(model_names, pipelines):

  # Time the training process
  training_time = timeit.timeit(
      stmt = '''pipeline.fit(df_train, y_train)''',
      globals = globals(), number = 1
      )

  # Print the training time for the current classifier
  print(f"Time taken to train {model_name}: {training_time:.2f} seconds")

  # Evaluate model
  accuracy, recall, precision, roc_auc_Score, fpr, tpr = evaluation_metrics(pipeline, df_val, y_val, model_name, labels)

  all_accuracy.append(accuracy)
  all_recall.append(recall)
  all_precision.append(precision)
  all_roc_auc.append(roc_auc_Score)
  all_fpr.append(fpr)
  all_tpr.append(tpr)
  print('===================================================================================')
```

```python
scores_table(model_names, all_accuracy, all_recall, all_roc_auc, 'Model Performance - Base')
```

```python
plot_barchart(model_names, all_accuracy, 'Accuracy', 'Accuracy Comparison - Base')
```

```python
plot_barchart(model_names, all_recall, 'Recall', 'Recall Comparison - Base')
```

```python
plot_roc_auc(model_names, all_fpr, all_tpr, all_roc_auc, 'ROC AUC CURVES - Base')
```

# TRAINING AND EVALUATION WITH OPTIMIZED HYPERPARAMETER

```python
grid_models = [
    LogisticRegression(random_state=42),
    SVC(random_state=42, probability=True),
    GaussianNB(),
    RandomForestClassifier(random_state=42),
    XGBClassifier(random_state=42)
    ]
```

```python
grid_pipelines = []

for model in grid_models:
  pipe = Pipeline([
      ('scaler', StandardScaler()),
      ('classifier', model)
      ])
  grid_pipelines.append(pipe)
```

```python
# Lists to store evaluation metrics across models
grid_accuracy = []
grid_recall = []
grid_precision = []
grid_roc_auc = []
grid_fpr = []
grid_tpr = []
```

```python
# Train and evaluate each model using the pipelines

for model_name, grid_pipeline in zip(model_names, grid_pipelines):

  # Time the training process
  training_time = timeit.timeit(
      stmt = '''grid_pipeline.fit(df_selected_oversampled, y_selected_oversampled)''',
      globals = globals(), number = 1
      )

  # Print the training time for the current classifier
  print(f"Time taken to train {model_name}: {training_time:.2f} seconds")

  # Evaluate model
  accuracy, recall, precision, roc_auc_Score, fpr, tpr = evaluation_metrics(grid_pipeline, df_val_selected_oversampled, y_val_selected_oversampled, model_name, labels)

  grid_accuracy.append(accuracy)
  grid_recall.append(recall)
  grid_precision.append(precision)
  grid_roc_auc.append(roc_auc_Score)
  grid_fpr.append(fpr)
  grid_tpr.append(tpr)
  print('===================================================================================')
```

```python
scores_table(model_names, grid_accuracy, grid_recall, grid_roc_auc, 'Model Performance - GA + SM + HO')
```

```python
plot_barchart(model_names, grid_accuracy, 'Accuracy', 'Accuracy Comparison - GA + SM + HO')
```

# ENSEMBLE LEARNING AND HYPERPARAMETER OPTIMIZATION

```python
lr = LogisticRegression(random_state=42),
svc = SVC(random_state=42, probability=True),
gnb = GaussianNB(),
rf = RandomForestClassifier(random_state=42),
xgb = XGBClassifier(random_state=42)

estimators = [
    ('lr', lr),
    ('svc', svc), 
    ('gnb', gnb), 
    ('rf', rf), 
    ('xgb', xgb)
]
```

```python
ensemble_clf = VotingClassifier(
    estimators = estimators,
    voting = 'soft',
    n_jobs = -1,
    verbose= True
)
```

```python
ensemble_clf.fit(df_selected_oversampled, y_selected_oversampled)
#ensemble_clf.predict(X)
accuracy, recall, precision, roc_auc_Score, fpr, tpr = evaluation_metrics(ensemble_clf, df_val_selected_oversampled, y_val_selected_oversampled, 'Ensemble Learning', )
```
